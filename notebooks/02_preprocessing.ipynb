## 3. Select Features
I'm selecting only features that are available for both on-market and off-market properties. Per guidance, ListPrice and OriginalListPrice are excluded because they introduce target leakage and won't be available for off-market properties.

```python
features = [
    'ClosePrice',       # target variable
    'LivingArea',       # interior square footage
    'BedroomsTotal',    # number of bedrooms
    'BathroomsTotalInteger',  # number of bathrooms
    'LotSizeAcres',     # lot size
    'YearBuilt',        # year home was built
    'GarageSpaces',     # number of garage spaces
    'PoolPrivateYN',    # whether home has a pool
    'Latitude',         # geographic location
    'Longitude',        # geographic location
    'CloseDate'         # needed for train/test split
]

df = df[features].copy()
print(f'Columns selected: {df.columns.tolist()}')
print(f'Shape: {df.shape}')
```

## 4. Remove Rows with Invalid ClosePrice
Only keeping rows where ClosePrice exists and is greater than zero since that's my target variable.

```python
before = len(df)
df = df[df['ClosePrice'].notna() & (df['ClosePrice'] > 0)]
print(f'Rows removed due to invalid ClosePrice: {before - len(df):,}')
print(f'Rows remaining: {len(df):,}')
```

## 5. Handle Missing Values
Since missing values are under 2% across all key columns (confirmed in Week 2 EDA), I'm dropping rows with any missing values. This keeps the data clean and avoids introducing artificial values through imputation.

```python
before = len(df)
df = df.dropna()
print(f'Rows dropped due to missing values: {before - len(df):,}')
print(f'Rows remaining: {len(df):,}')
```

## 6. Encode Categorical Columns
PoolPrivateYN is a True/False column — converting it to 1/0 so the model can use it numerically.

```python
df['PoolPrivateYN'] = df['PoolPrivateYN'].map({True: 1, False: 0, 'True': 1, 'False': 0})

print('PoolPrivateYN value counts:')
print(df['PoolPrivateYN'].value_counts())
```

## 7. Add Property Age Feature
Instead of using YearBuilt directly, I'm engineering a more meaningful feature: how old the home was at the time of sale.

```python
df['CloseDate'] = pd.to_datetime(df['CloseDate'], errors='coerce')
df['PropertyAge'] = df['CloseDate'].dt.year - df['YearBuilt']

# drop any rows where PropertyAge came out negative or unrealistic
df = df[df['PropertyAge'] >= 0]

print(df['PropertyAge'].describe().round(1))
```

## 8. Train/Test Split
Per instructions:
- **Test set** = most recent full month available = May 2026
- **Training set** = 12 months immediately before that (May 2025 – April 2026)

I'm starting with 12 months as my training window and will experiment with different window sizes in later weeks to find the optimal value.

```python
df['YearMonth'] = df['CloseDate'].dt.to_period('M')

# test set = May 2026
test = df[df['YearMonth'] == '2026-05'].copy()

# training set = 12 months before test (May 2025 to April 2026)
train = df[
    (df['YearMonth'] >= '2025-05') &
    (df['YearMonth'] <= '2026-04')
].copy()

print(f'Training set: {len(train):,} rows')
print(f'Test set:     {len(test):,} rows')
print(f'Training window: {train["YearMonth"].min()} to {train["YearMonth"].max()}')
print(f'Test window:     {test["YearMonth"].min()} to {test["YearMonth"].max()}')
```

## 9. Define Features and Target
Dropping CloseDate and YearMonth since they were only needed for the split — not actual model inputs.

```python
feature_cols = [
    'LivingArea', 'BedroomsTotal', 'BathroomsTotalInteger',
    'LotSizeAcres', 'PropertyAge', 'GarageSpaces',
    'PoolPrivateYN', 'Latitude', 'Longitude'
]

X_train = train[feature_cols]
y_train = train['ClosePrice']

X_test = test[feature_cols]
y_test = test['ClosePrice']

print(f'X_train shape: {X_train.shape}')
print(f'X_test shape:  {X_test.shape}')
```

## 11. Preprocessing Summary

**Decisions made:**
- Excluded `ListPrice` and `OriginalListPrice` to avoid target leakage and ensure the model works for off-market properties
- Dropped rows with missing values — missing data was under 2% so minimal data loss
- Encoded `PoolPrivateYN` as 1/0
- Engineered `PropertyAge` from `YearBuilt` and `CloseDate`
- Test set = May 2026 (most recent full month)
- Training window = 12 months (May 2025 – April 2026) — will experiment with other window sizes in later weeks

### Why so many rows dropped?
In Week 2 each column individually had under 2% missing values. However when dropping 
rows, a row is removed if ANY of its columns has a missing value. With 9 feature columns, 
these small percentages stack up across columns, resulting in 15% of rows being removed. 
We still have 340,126 clean rows remaining which is more than enough for modeling.
